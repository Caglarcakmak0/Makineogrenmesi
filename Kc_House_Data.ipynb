```python
X= df2.drop(["price"],axis=1)
y=df2["price"]
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
import pandas as pd

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=35)

def eniyimodelibul(X, y):
    algoritmalar = {
        "linear_regression": {
            "model": LinearRegression(),
            "parametreler": {
                'copy_X': [True, False],
                'fit_intercept': [True, False],
                'n_jobs': [None, 1, 2, 3, 4], 
                'positive': [False, True]
            }
        },
        "lasso": {
            "model": Lasso(),
            "parametreler": {
                'alpha': [1, 2],
                'selection': ['random', 'cyclic']
            }
        },
        "decision_tree": {
            "model": DecisionTreeRegressor(),
            "parametreler": {
                'criterion': ['friedman_mse', 'poisson', 'squared_error', 'absolute_error'],
                'splitter': ['best', 'random']
            }
        }
    }

    skorlar = []  
    for algoritma, models in algoritmalar.items():
        gs = GridSearchCV(models["model"], models["parametreler"], cv=5, return_train_score=False)
        gs.fit(X, y)
        skorlar.append({
            'model': algoritma,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })
    return pd.DataFrame(skorlar, columns=['model', 'best_score', 'best_params'])

eniyimodelibul(X, y)
```
